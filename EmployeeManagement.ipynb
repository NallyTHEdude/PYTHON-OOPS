```python
class Employee:
    def __init__(self,name:str,age:int,salary:float,id:int) -> None:
        self.name = name
        self.age = age
        self.salary = salary
        self.id = id
    
    def __str__(self) -> str:
        return f"Name:{self.name},Age:{self.age},Sal:{self.salary},ID: {self.id}"



#-------------------------------------------------EMPLOYEE MANAGER CLASS -------------------------------------------------#
class EmployeeManager:
    def __init__(self) -> None:
        self.employeeList = []
    
    def add_employee(self):
        print("Enter employee details\n")
        name = input("Enter employee name: ")
        age = int(input("Enter employee age: "))
        salary = float(input("Enter employee salary: "))
        empId = int(input("Enter employee ID: "))
        emp = Employee(name,age,salary,empId)
        self.employeeList.append(emp)
    
    def print_emp(self):
        if len(self.employeeList) == 0:
            print("!!!---No employees Present, add employees---!!!")
        else:
            for emp in self.employeeList:
                print(emp)
    
    def delete_employees_with_age(self,startAge:int,endAge:int):
        if len(self.employeeList) ==0:
            print("!!!---No employees Present, add employees---!!!")
        else:
            for i in self.employeeList:
                if startAge < i.age < endAge:
                    print(f"Successfully removed employee \"({i.name},{i.id})\".....")
                    self.employeeList.remove(i)
    
    def find_employee_with_id(self,id):
        if len(self.employeeList) ==0:
            print("!!!---No employees Present, add employees---!!!")
        else:
            for i in self.employeeList:
                if i.id == id:
                    return i
        return None
    
    def update_employee_salary(self,id:int,newSalary:float):
        emp = self.find_employee_with_id(id)
        if emp is None:
            print("!!!---No employees with this ID present---!!!")
        else:
            emp.salary = newSalary
            
            

#-------------------------------------------------EMPLOYEE MANAGER CLASS -------------------------------------------------#
class EmployeeFrontend:
    def __init__(self) -> None:
        self.empManager = EmployeeManager()
    
    def print_menu(self):
        pass
        #HAVE TO FINISH THIS FUNCTION SOON ENOUGH
        
```
